```python
stocks_df = pd.read_csv("stocks_data.csv", header=[0,1], index_col=[0], parse_dates=[0])
```

```python
stocks_df
```

```python
stocks_df.columns
```

```python
# Convertendo index multiplo para tupla
stocks_df.columns = stocks_df.columns.to_flat_index()

stocks_df.columns
```

```python
stocks_df
```

```python
# Voltando para o formato original
stocks_df.columns = pd.MultiIndex.from_tuples(stocks_df.columns)

stocks_df.columns
```

```python
stocks_df.describe()
```

```python
close = stocks_df["Close"].copy()

close
```

```python
%matplotlib inline
plt.style.use("seaborn")
```

```python
close.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
close.iloc[0,0]
```

```python
close.AAPL.div(close.iloc[0,0]).mul(100)
```

```python
normalized_close = close.div(close.iloc[0]).mul(100)
```

```python
normalized_close.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
aapl = close.AAPL.copy().to_frame()

aapl
```

```python
aapl["shifted"] = aapl.shift(periods=1)

aapl
```

```python
aapl["Diff"] = aapl.AAPL.diff(periods=1)

aapl
```

```python
aapl["% Change"] = aapl.AAPL.pct_change(periods=1).mul(100)

aapl
```

```python
# Pega o valor do Ãºltimo valor do mÃªs
aapl.AAPL.resample("M").last()
```

```python
# Pega o valor do Ãºltimo valor do Ãºltimo dia Ãºtil do mÃªs
aapl.AAPL.resample("BM").last()
```

```python
aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

```python
ret = aapl.AAPL.pct_change().dropna()

ret
```

```python
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

```python
daily_mean_ret = ret.mean()*100
daily_mean_ret
```

```python
var_daily = ret.var()
var_daily
```
